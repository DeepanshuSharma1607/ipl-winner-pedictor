```python
df=df.drop_duplicates('match_id',keep='first').reset_index(drop=True)
```

```python
team_mapping = {
    'Delhi Capitals': 'DC',
    'Delhi Daredevils': 'DC',

    'Punjab Kings': 'PBKS',
    'Kings XI Punjab': 'PBKS',

    'Royal Challengers Bangalore': 'RCB',
    'Royal Challengers Bengaluru': 'RCB',

    'Kolkata Knight Riders': 'KKR',
    'Chennai Super Kings': 'CSK',
    'Rajasthan Royals': 'RR',
    'Mumbai Indians': 'MI',
    'Sunrisers Hyderabad': 'SRH',
    'Lucknow Super Giants': 'LSG',
    'Gujarat Titans': 'GT',

    'Deccan Chargers': 'DC_OLD',
    'Kochi Tuskers Kerala': 'KTK',
    'Pune Warriors': 'PWI',
    'Rising Pune Supergiants': 'RPS',
    'Rising Pune Supergiant': 'RPS',
    'Gujarat Lions': 'GL',

}


df['batting_team'] = df['batting_team'].replace(team_mapping)
df['bowling_team'] = df['bowling_team'].replace(team_mapping)

df['match_won_by'] = df['match_won_by'].replace(team_mapping)
df['toss_winner'] = df['toss_winner'].replace(team_mapping)
```

```python
df = df[df['match_won_by'] != 'Unknown'].reset_index(drop=True)
```

```python
df['venue']=(df['venue']
             .str.split(',').str[0]
             .str.strip()
             .str.lower()
             .str.replace(' ', '_'))
```

```python
all_teams=df['batting_team'].unique()
team_scores={team:1500 for team in all_teams}
bat_perf=[]
bowl_perf=[]
for index , row in df.iterrows():
    bat_team=row['batting_team']
    bowl_team=row['bowling_team']
    winner=row['match_won_by']

    bat_perf.append(team_scores[bat_team])
    bowl_perf.append(team_scores[bowl_team])

    if winner==bat_team:
      team_scores[bat_team]+=25
      team_scores[bowl_team]-=25
    else:
      team_scores[bat_team]-=25
      team_scores[bowl_team]+=25
df['batting_team_perf']=bat_perf
df['bowling_team_perf']=bowl_perf
```

```python
map={
    'batting_team':'team_1',
    'bowling_team':'team_2',
    'match_won_by':'winner'
}
df=df.rename(columns=map)
```

```python
df['is_team_1_winner']=(df['team_1']==df['winner']).astype(int)
df['is_team_2_winner']=(df['team_2']==df['winner']).astype(int)

win_counts ={team : 0 for team in pd.concat([df['team_1'],df['team_2']]).unique()}
t1_counts=[]
t2_counts=[]

for i, row in df.iterrows():
  t1_counts.append(win_counts[row['team_1']])
  t2_counts.append(win_counts[row['team_2']])

  win_counts[row['winner']]+=1

df['team_1_win_count']=t1_counts
df['team_2_win_count']=t2_counts
df.drop(columns=['is_team_1_winner', 'is_team_2_winner'], inplace=True)
```

```python
x_train=df[df['year']<=2024].copy()
x_test=df[df['year']>2024].copy()
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
y_train=x_train['winner']
x_train=x_train.drop(columns=['match_id','winner','year'])
y_test=x_test['winner']
x_test=x_test.drop(columns=['match_id','winner','year'])
```

```python
from catboost import CatBoostClassifier
cat_features=['team_1', 'team_2', 'toss_winner', 'toss_decision', 'venue','stage']
clf = CatBoostClassifier(
    iterations=500,
    learning_rate=0.05, # Lowering this also helps (see point 2)
    depth=4,            # Reducing depth (see point 3)
    l2_leaf_reg=10,     # Higher value = more regularization
    loss_function='MultiClass',
    verbose=100
)
clf.fit(
    x_train, y_train,
    eval_set=(x_test, y_test), # Monitor the test set
    early_stopping_rounds=50,  # Stop if no improvement for 50 rounds
    cat_features=cat_features
)
```

```python
from sklearn.metrics import accuracy_score
pred_test=clf.predict(x_test)
pred_train=clf.predict(x_train)
print(accuracy_score(y_train,pred_train))
print(accuracy_score(y_test,pred_test))
```

```python
import pandas as pd
feat_imp = pd.Series(clf.get_feature_importance(), index=x_train.columns)
print(feat_imp.sort_values(ascending=False))
```
